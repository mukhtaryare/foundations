## Scraping

### Loop through each `tr` and print each person's name

You'll get an error because the first one doesn't have a name. How do you make that not happen? I'm happy to help if you ask on the board.

```python
name = doc.find_all('tr')
for item in name:
    if item.find('span') != None:
        print(item.find('span').text)
```

## Loop through each `tr`, printing each violation description

- TIP: What is the container tag name for it?
- TIP: You'll get an error even if you're ALMOST right - which row is causing the problem?

```python
name = doc.find_all('tr')
for item in name[1:]: # get the second tr 
    print(item.find_all('td')[2].text) # thrid column
```

## Loop through each `tr`, printing the complaint number

- TIP: It should be the last piece of the fist `td`

```python
name = doc.find_all('tr')
for item in name[1:]:
    print(item.find_all(class_='results_text')[-2].string) # second to last (last is date)
```

## Saving the results

### Loop through each `tr` to create a list of dictionaries

Each dictionary must contain

- Person's name
- Violation description
- Violation number

Create a new dictionary for each `tr` (except the header).

```python
# Start with an empty list

texas_barber_violations = []

name = doc.find_all('tr')
for item in name:
    current = {}
    
    if item.find('span') != None:
        current['name'] = item.find('span').text
        current['violation_description'] = item.find_all('td')[2].text
        current['complaint_numbers'] = item.find_all(class_='results_text')[-2].string
        
        texas_barber_violations.append(current)
```

### Save that to a CSV

```python
import pandas as pd
df = pd.DataFrame(texas_barber_violations)
df.to_csv("texas_barber_violations.csv", index=False)
```
